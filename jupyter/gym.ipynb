```python
import matplotlib.pyplot as plot
```

```python
max = 1000
max_steps = 900
vals = list()
lr = 0.00002
for i in range(max):
    lr_scale = 1.0
    if i < max_steps:
        lr_scale = min(1., float(i + 1) / max_steps)
    vals.append(lr_scale*lr)

```

```python
plot.plot(vals)
```
